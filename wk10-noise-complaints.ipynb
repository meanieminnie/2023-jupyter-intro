```python
# import median income data
mi = pd.read_csv("ACS-median-income-nyc-zipcode.csv")
mi
```

```python
# filter large median census data to smaller dataset
midf = mi.filter(["NAME", "S1903_C01_012E"])
midf
```

```python
#  rename columns
midf = midf.rename(columns={"NAME": "zip", "S1903_C01_012E": "median-income"})
midf
```

```python
midf.info()
```

```python
# convert median incoem data from object to float(integer)
midf["median-income"] = pd.to_numeric(midf["median-income"], errors = "coerce")

midf.info()
```

```python
# remove the extra parts from zip code
midf["zip"] = midf["zip"].str.replace("ZCTA5 ","")
midf
                        
```

```python
midf.to_csv("median-income-2021.csv", index = False)
```
